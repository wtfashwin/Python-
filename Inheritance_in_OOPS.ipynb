## Multiple Inheritance


> 2 parent class having a single child


```python
# class class_parent1:
#   body
# class class_parent2:
#   body
# class child(class_parent1, class_parent2):
#   body
```

```python
class mom:
  momname = ''
  def mommy(self):
    print(self.momname)

class dad:
  dadname = ''
  def daddy(self):
    print(self.dadname)

class child(mom, dad):
  def parent(self):
    print('Dad Name:', self.dadname)
    print('Mom Name:', self.momname)

c1 = child()
c1.momname = 'Arch'
c1.dadname = 'Dee'
c1.parent()
```

##Multi Level Inheritance

A --> B --> C

image.png

```python
class grand:
  def grandname(self, name):
    self.grandname = name
class fath(grand):
  def fname(self, name):
    self.fname = name
class son(fath):
  def son(self, name):
    self.son = name
  def printname(self):
    print('Grand Father Name:', self.grandname)
    print('Father Name:', self.fname)
    print('Son Name:', self.son)
s = son()
s.grandname('A')
s.fname('B')
s.son('C')
s.printname()
```

```python
class grandfather:
    def __init__(self, grandfathername):
        self.grandfathername = grandfathername

class father(grandfather):
    def __init__(self, fathername, grandfathername):
        super().__init__(grandfathername)
        self.fathername = fathername

class son(father):
    def __init__(self, sonname, fathername, grandfathername):
        super().__init__(fathername, grandfathername)
        self.sonname = sonname

    def print_names(self):
        print("Son's Name:", self.sonname)
        print("Father's Name:", self.fathername)
        print("Grandfather's Name:", self.grandfathername)
son1 = son("C", "B", "A")
son1.print_names()
```

```python
class grandfather:
  def gname(self, grandfather_name):
    self.grandfather_name = grandfather_name

class father(grandfather):
  def fname(self, father_name):
    self.father_name = father_name

class son(father):
  def so(self, son_name):
    self.son_name = son_name

  def print_all_names(self):
    print("Grandfather's Name:", self.grandfather_name)
    print("Father's Name:", self.father_name)
    print("Son's Name:", self.son_name)

my_family = son()

my_family.gname("X")
my_family.fname("Y")
my_family.so("Z")

my_family.print_all_names()
```

```python
class Grandfather:
    def grandfather_name(self, name):
        self.grandfathername = name

class Father(Grandfather):
    def father_name(self, name):
        self.fathername = name

class Son(Father):
    def son_name(self, name):
        self.sonname = name
    def print_names(self):
        print("Grandfather name:", self.grandfathername)
        print("Father name:", self.fathername)
        print("Son name:", self.sonname)

s1 = Son()
s1.grandfather_name("A (Grandfather)")
s1.father_name("B (Father)")
s1.son_name("C (Son)")
s1.print_names()
```

##Hierarchical Inheritance


image.png

```python
class animal:
  def speak(self):
    print('Animals Can Make Sounds')

class dog(animal):
  def bark(self):
    print('Dog can Bark')

class cat(animal):
  def meow(self):
    print('Cats Meow')

class lion(animal):
  def roar(self):
    print('Lions Roar')

d = dog()
c = cat()
l = lion()

d.speak()
d.bark()
c.speak()
c.meow()
l.speak()
l.roar()
```

## Hybrid Inheritance

- more than one types of inheritance used

image.png

```python
class person:
  def details(self):
    print(' I am a Person')    ##Single Inh

class emp(person):
  def work(self):
    print('I work')         #Hierarchical Inh

class student(person):
  def study(self):
    print('I study')

class manager(emp, student):    #Multiple Inh
  def manage(self):
    print('I Manage')
```

```python
m1 = manager()
m1.study()
```
